```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from mpltable import ColumnDefinition, Table
```

```python
d = pd.DataFrame(np.random.random((10, 5)), columns=["A", "B", "C", "D", "E"]).round(2)

fig, ax = plt.subplots(figsize=(5, 8))

tab = Table(d)

plt.show()

fig.savefig("images/basic_table.png")
```

```python
c = tab.get_column("A")
```

```python
tab.textprops
```

```python
fig, ax = plt.subplots(figsize=(5, 8))
tab = Table(
    d,
    column_definitions=[
        ColumnDefinition(name="A", title="Title A", textprops={"ha": "center"}, width=2)
    ],
)
plt.show()
```

```python
d = pd.DataFrame(np.random.random((10, 5)), columns=["A", "B", "C", "D", "E"]).round(2)

fig, ax = plt.subplots(figsize=(5, 8))

tab = Table(d, row_dividers=False).set_alternating_row_colors("#f2f9fc", "#e0f6ff")

plt.show()

fig.savefig("images/alternating_row_color.png")
```
